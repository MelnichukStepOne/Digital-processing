```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import io
```

#Основне завдання

```python
url_color = "https://klike.net/uploads/posts/2019-05/1556708011_11.jpg"
url_gray ="https://klike.net/uploads/posts/2020-10/1602584782_12.jpg"

img_color_BGR = io.imread(url_color)
img_gray = io.imread(url_gray)

cv2_imshow(img_color_BGR) 
cv2_imshow(img_gray) 
```

```python
img_color_RGB = cv2.cvtColor(img_color_BGR, cv2.COLOR_BGR2RGB) 
cv2_imshow(img_color_RGB)
```

```python
img_color_HSV = cv2.cvtColor(img_color_BGR, cv2.COLOR_BGR2HSV) 
cv2_imshow(img_color_HSV)
```

```python
#RGB

r = img_color_RGB.copy()
r[:, :, 0] = 0
r[:, :, 1] = 0

g = img_color_RGB.copy()
g[:, :, 0] = 0
g[:, :, 2] = 0

b = img_color_RGB.copy()
b[:, :, 1] = 0
b[:, :, 2] = 0

cv2_imshow( cv2.hconcat( (r,g,b) ) )
```

```python
#HSV

h = img_color_HSV.copy()
h[:,:,0]=0
h[:,:,1]=0

s = img_color_HSV.copy()
s[:,:,0]=0
s[:,:,2]=0

v = img_color_HSV.copy()
v[:,:,1]=0
v[:,:,2]=0

cv2_imshow( cv2.hconcat( (h,s,v) ) )
```

```python
# Region of interest
ROI_height = int(img_gray.shape[0]/2)
ROI_width = int(img_gray.shape[1]/2)

img_gray_roi = img_gray[0 : ROI_height, 0 : ROI_width]

cv2_imshow( img_gray_roi )
```

```python
cv2.imwrite("Gray_ROI.png",img_gray_roi)
```
